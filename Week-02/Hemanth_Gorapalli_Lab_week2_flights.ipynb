```python
import pandas as pd
import numpy as np


flights = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/nyc_flights.csv')

weather = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/relational_data/nyc_weather.csv')

# example merge:
# flights.merge(weather, on= ['year', 'month', 'day', 'hour', 'origin'])
```

```python
# Define a custom function to get the first and last flights
def get_first_last_flights(group):
    return pd.Series({'first_flight': group['dep_time'].min(), 'last_flight': group['dep_time'].max()})

# Calculate first and last flights each day
first_last_flights = flights.groupby(['year', 'month', 'day']).apply(get_first_last_flights)
first_last_flights
```

```python
# Count canceled flights per carrier and origin
cancelled_flights = flights[flights['dep_time'].isnull()].groupby(['carrier', 'origin']).size()
cancelled_flights
```

```python
worst_dep_delay_carrier = flights.groupby('carrier')['dep_delay'].mean().sort_values(ascending=False)
worst_dep_delay_carrier
```

```python
# Count non-null departure delays per carrier
non_null_counts = flights.groupby('carrier')['dep_delay'].apply(lambda x: x.notnull().sum())

# Calculate total departure delay per carrier
total_dep_delay_carrier = flights.groupby('carrier')['dep_delay'].sum()

# Calculate average departure delay per carrier
average_dep_delay_carrier = total_dep_delay_carrier / non_null_counts

# Find carrier with the worst departure delays
worst_dep_delay_carrier = average_dep_delay_carrier.idxmax()

print("Carrier with the worst departure delays:", worst_dep_delay_carrier)
```

```python
# Find the plane with the worst on-time record based on departure delays
worst_on_time_plane = flights.groupby('tailnum')['dep_delay'].mean().nlargest(1)

print("Plane with the worst on-time record:", worst_on_time_plane.index[0])
```

```python
# Function to calculate the number of flights before a >1 hr delay per plane
def count_flights_before_delay(group):
    group_sorted = group.sort_values('dep_delay')
    group['flights_before_delay'] = range(len(group_sorted))
    return group

# Calculate the number of flights before a >1 hr delay per plane
flights_before_delay = flights.groupby('tailnum', group_keys=False).apply(count_flights_before_delay)

# Filter flights with >1 hr delay
flights_before_delay = flights_before_delay[flights_before_delay['dep_delay'] > 60]

# Select relevant columns for structured representation
result_df = flights_before_delay[['tailnum', 'flight', 'dep_delay', 'flights_before_delay']]

# Reset index for a cleaner look
result_df.reset_index(drop=True, inplace=True)

# Display the result dataframe
result_df.head()

```
